# Filtering and Sorting Data: Chipotle

Step 1. Import the necessary libraries

```python
import pandas as pd
```

Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipotle = pd.read_csv(url, sep = '\t')
#print(chipotle)
```

```python
prices = [float(value) for value in chipotle.item_price]

# reassign the column with the cleaned prices
chipotle.item_price = prices

# delete the duplicates in item_name and quantity
chipotle_filtered = chipotle.drop_duplicates(['item_name','quantity','choice_description'])
#print(chipotle_filtered)

# select only the products with quantity equals to 1
chipotle_one_prod = chipotle_filtered[chipotle_filtered.quantity == 1]
#print(chipotle_one_prod)

chipotle.query('item_price > 10')   #.item_name  #.nunique()
```

Step 5. What is the price of each item?
print a data frame with only two columns item_name and item_price

Step 6. Sort by the name of the item

```python
#chipotle.item_name.sort_values()

chipotle.sort_values(by = "item_name")
```

Step 7. What was the quantity of the most expensive item ordered?

```python
chipotle.sort_values(by = "item_price", ascending = False)   #.head(1)
```

Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipotle_salad = chipotle[chipotle.item_name == "Veggie Salad Bowl"]
print(chipotle_salad)

len(chipotle_salad)
```

Step 9. How many times did someone order more than one Canned Soda?

```python
more_than_one_soda = chipotle[(chipotle.item_name == "Canned Soda") & (chipotle.quantity > 1)]
print(more_than_one_soda)

len(more_than_one_soda)
```
