# MatPlotLib 

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.__version__    # Check version
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), label='line 1')
plt.plot(x, np.cos(x), label='line 2')
plt.legend()
plt.show()
```

```python
#%matplotlib  # Give an interactive plot
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-g')
plt.plot(x, np.cos(x), '--')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```
